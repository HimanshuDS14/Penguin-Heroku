```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv("Penguins_cleaned.csv")
data.head(10)
```

```python
data.tail()
```

```python
target = ['species']
encode = ['sex' , 'island']
```

```python
for col in encode:
    dummy  = pd.get_dummies(data[col] , prefix = col)
    data = pd.concat([data , dummy] , axis = 1)
    del data[col]
```

```python
data.head(5)
```

```python
target_mapper = { 'Adelie' : 0 , 'Chinstrap' : 1 , 'Gentoo' : 2 }
```

```python
def target_encode(val):
    return target_mapper[val]
```

```python
data['species'] = data['species'].apply(target_encode)
```

```python
data.head(10)
```

```python
x = data.drop('species' , axis=1)
y = data['species']
```

```python
model = RandomForestClassifier()
model.fit(x,y)
```

```python
import pickle

pickle.dump(model , open("Penguin_model_01.pkl" , 'wb'))
```
